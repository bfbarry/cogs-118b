```python
%matplotlib inline

import numpy as np
from scipy.io import loadmat

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D

```

# Question 6

```python
p1 = loadmat('hw1p1.mat')
p1_data = p1['hw1p1data']
p1_x = p1_data.T[0]
p1_y = p1_data.T[1]
```

```python
plt.plot(p1_x, p1_y)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('P1 Data')
plt.text(5, 60, 'plt.plot(p1_x, p1_y)')
plt.show()
```

# Question 7

```python
p2 = loadmat('hw1p2.mat')
p2_data = p2['hw1p2data']
p2_x = p2_data.T[0]
p2_y = p2_data.T[1]
p2_z = p2_data.T[2]
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
#c =p2_z
ax.plot_trisurf(p2_x, p2_y, p2_z, cmap='plasma',linewidth=0.2, antialiased=True)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.title('P2 Data')
plt.show()
```
